```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("kc_house_data.csv")
df.head()
```

##### 1.2 
Изучите стоимости недвижимости

```python
df['price'].describe()
graf = plt.figure(figsize=[15,10])
graf1 = graf.add_subplot(1,2,2)
graf1.hist(df['price'], bins=100)
graf1.set(xlabel='Стоимость', ylabel='Количество', title='Цены на дома')
plt.show()
```

Большая часть домов стоит до 1млн

##### 1.3 
Изучите распределение квадратуры жилой

```python
df['sqft_living'].describe()
graf = plt.figure(figsize=[30,10])
graf2 = graf.add_subplot(1,2,2)
graf2.hist(df['sqft_living'], bins=200)
graf2.set(xlabel='Жилая площадь', ylabel='Частота', title='Распределение жилой площади')
plt.show()
```

Большинство недвижимости имеет площадь от 1500 до 4000.

##### 2.1.4
Изучите распределение года постройки

```python
data = df.groupby('yr_built')['price'].sum().reset_index()
plt.figure(figsize=(10,8))
plt.scatter(data['yr_built'],data['price'])
plt.title('Зависимость цены от года постройки')
plt.xlabel('Год постройки')
plt.ylabel('Цена');
```
